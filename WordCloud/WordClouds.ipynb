```python
RWC['Russian'] = RWC['Russian'].replace(regex=[","], value=' ')
RWC['Russian'] = RWC['Russian'].replace(regex=["'"], value=' ')
text = ' '.join(RWC.Russian)
text = text.replace('[', "")
text = text.replace(']', "")
text = text.replace('   ', " ")
text = text.replace('  ', " ")
text = text.replace('  ', " ")
```

```python
FBWC['FB API'] = FBWC['FB API'].replace(regex=[","], value=' ')
FBWC['FB API'] = FBWC['FB API'].replace(regex=["'"], value=' ')
text2 = ' '.join(FBWC['FB API'])
text2 = text.replace('[', "")
text2 = text.replace(']', "")
text2 = text2.replace('   ', " ")
text2 = text2.replace('  ', " ")
text2 = text2.replace('  ', " ")
```

```python
mask2 = np.array(Image.open("Image8.png"))
```

```python
def generate_wordcloud(text2, mask2):
    word_cloud = WordCloud(width = 1600, height = 800, background_color='white', stopwords=STOPWORDS, mask=mask2).generate(text2)
    plt.figure(figsize=(10,5),facecolor = 'k', edgecolor='blue')
    plt.imshow(word_cloud)
    plt.axis('off')
    plt.title("FB API Word Cloud")
    plt.tight_layout(pad=0)
    plt.savefig('FB_Word_Cloud.png', figsize = (1200,400), dpi=600)
    plt.show()
```

```python
generate_wordcloud(text2, mask2)
```
