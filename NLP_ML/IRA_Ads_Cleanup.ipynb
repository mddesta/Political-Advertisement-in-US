```python
import pandas as pd

```

```python
all_ads = pd.read_excel("all_fb_ads.xlsx")
```

```python
all_ads.head()
```

```python
all_ads = all_ads.dropna(subset=['adText'])
all_ads = all_ads.drop('groups',axis=1)
```

```python
all_ads.head()
all_ads.dtypes
```

```python
cleaned_ads = all_ads
cleaned_ads["fix_date"] = cleaned_ads["adCreation"].astype(str)
cleaned_ads["length"] = cleaned_ads["fix_date"].str.len()
```

```python
cleaned_ads.head()
```

```python
cleaned_data = cleaned_ads
cleaned_data = cleaned_data.drop(cleaned_data[cleaned_data.length > 25].index)
```

```python
print(len(cleaned_data))
```

```python
cleaned_data.to_csv("cleaned_data.csv")
```
