```python
import os
import pandas as pd
import numpy as np
import glob
from datetime import datetime
import dateutil
```

```python
path = "Output"
all_files = glob.glob(path + "/*.csv")
li=[]

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

ads_df = pd.concat(li, axis=0, ignore_index=True, sort=True)
ads_df
```
