```python
# Import Dependencies
import pandas as pd
import requests
import json
import numpy as np
```

```python
# Create a reference the CSV file desired
file_path = "Russian_FB_Ads.xlsx"

# Read the CSV into a Pandas DataFrame
RussianAds = pd.read_excel(file_path)

# Print the first five rows of data to the screen
RussianAds.head()
```

```python
RussianAds.dropna(subset=['adStates'],inplace=True)
RussianAds.head()
```

```python
State = RussianAds['adStates'].str.split(' ',expand=True).stack()

#join the new series (after setting a name and index) back to the dataframe
RussianAds1= RussianAds.join(pd.Series(index=State.index.droplevel(1), data=State.values, name='State'))
RussianAds2 = RussianAds1[RussianAds1.State != "New"]
RussianAds4 = RussianAds2[RussianAds2.State != "Carolina"]
RussianAds3 = RussianAds4[RussianAds4.State != "Washington"]

```

```python
RussianAds3.State = RussianAds3.State.str.replace("York", "New York")
RussianAds3.State = RussianAds3.State.str.replace("Columbia", "Washington D.C.")
RussianAds3.State = RussianAds3.State.str.replace("Hampshire", "New Hampshire")
```

```python
RussianAds3.State = RussianAds3.State.str.replace("Hampshire", "New Hampshire")
RussianAds3.State = RussianAds3.State.str.replace("Jersey", "New Jersey")
RussianAds3.State = RussianAds3.State.str.replace("Mexico", "New Mexico")
RussianAds3.State = RussianAds3.State.str.replace("North", "North Carolina")
RussianAds3.State = RussianAds3.State.str.replace("South", "South Carolina")

Ads = RussianAds3
```
