```python
# Import Dependencies
import pandas as pd
import requests
import json
```

```python
# Create a reference the CSV file desired
file_path = "Russian_FB_Ads.xlsx"

# Read the CSV into a Pandas DataFrame
RussianAds = pd.read_excel(file_path)

# Print the first five rows of data to the screen
RussianAds.head()
```

```python
RussianAds.dropna(subset=['adStates'],inplace=True)
RussianAds.head()
```

```python
State = RussianAds.adStates.str.split(' ',expand=True).stack()
#join the new series (after setting a name and index) back to the dataframe
Ads= RussianAds.join(pd.Series(index=State.index.droplevel(1), data=State.values, name='State'))
Ads.head()
```
